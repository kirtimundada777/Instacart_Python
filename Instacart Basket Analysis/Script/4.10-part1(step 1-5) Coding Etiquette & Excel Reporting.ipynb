##  Create a profiling variable based on age, income, certain goods in the “department_id” column, and number of dependents.

### Creating age groups

```python
df_active_customers['age'].describe()
```

based on the above statistics for the age group , I created these age groups: 18 to 30 years old, 31 to 45 years old, 46 to 60 years old, 61 to 81 years old.

```python
df_active_customers.loc[(df_active_customers['age'] >= 18) & (df_active_customers['age'] <= 30), 'age_group'] = '18-30 years old'
```

```python
df_active_customers.loc[(df_active_customers['age'] >= 31) & (df_active_customers['age'] <= 45), 'age_group'] = '31-45 years old'
```

```python
df_active_customers.loc[(df_active_customers['age'] >= 46) & (df_active_customers['age'] <= 60), 'age_group'] = '46-60 years old'
```

```python
df_active_customers.loc[(df_active_customers['age'] >= 61) & (df_active_customers['age'] <= 81), 'age_group'] = '61-81 years old'
```

```python
df_active_customers.head()
```

```python
df_active_customers['age_group'].value_counts(dropna=False)
```

### Creating income groups

```python
df_active_customers['income'].describe()
```

### From the above statistics  for the income column, I created these age groups: less than 67292 - 'Low Income',  more than  67292  and less than 128102 -'Middle Income' and  more than 128102 - high income

```python
df_active_customers.loc[df_active_customers['income'] <= 67292, 'income_group'] = 'Low Income'
```

```python
df_active_customers.loc[(df_active_customers['income'] > 67292) & (df_active_customers['income'] <= 128102), 'income_group'] = 'Middle Income'
```

```python
df_active_customers.loc[df_active_customers['income'] > 128102, 'income_group'] = 'High Income'
```

```python
df_active_customers.head()
```

```python
df_active_customers['income_group'].value_counts(dropna=False)
```
