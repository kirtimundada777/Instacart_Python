## Step5: Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day.

```python
np.random.seed(4)
dev = np.random.rand(len(df_ords_prods_merge)) <= 0.7
```

```python
big = df_ords_prods_merge[dev]
```

```python
small = df_ords_prods_merge[~dev] 
```

```python
len(df_ords_prods_merge)
```

```python
len(small)+len(big)
```

```python
df = small[['prices', 'order_hour_of_day']]
```

```python
#create a line graph
line_price_hour = sns.lineplot(data = df, x = 'order_hour_of_day',y = 'prices')
```

There's no significant correlation between prices and order hour of day

```python
#Export visualization
line_price_hour.figure.savefig(os.path.join(path, 'Analysis','Visualizations', 'line_price_hour.png'))
```
